```python
import pandas as pd
```

```python
data = pd.read_csv("data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
data.head()
```

```python
data.drop("customerID", axis = 1, inplace = True)
data.sample(5)
```

```python
data1 = data[data.TotalCharges != " "]
data1.TotalCharges.isnull()
```

```python
# data1.loc[:,"TotalCharges"] = pd.to_numeric(data1['TotalCharges'])
data1.TotalCharges = pd.to_numeric(data1.TotalCharges)
```

```python
data1.dtypes
```

```python
def unique_val(data1):
    for column in data1:
        if data1[column].dtypes =='object':
            print(f'{column} :{data1[column].unique()}')
```

```python
unique_val(data1)
```

```python
data1.replace("No internet service",'No',inplace=True)
data1.replace("No phone service",'No',inplace=True)
```

```python
unique_val(data1)
```

```python
yes_no = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
for word in yes_no:
    data1[word].replace({'Yes' : 1,'No' :0},inplace=True)
```

```python
for col in data1:
    print(f'{col} :{data1[col].unique()}')
```

```python
data1.replace({'Female': 1,'Male': 0},inplace=True)
```

```python
data1['gender'].unique()
```

```python
for col in data1:
    print(f'{col}: {data1[col].unique()}')
```

```python
data2 = pd.get_dummies(data=data1,columns=['InternetService','Contract','PaymentMethod'],dtype=int)
data2
```

```python
scale_col = ['tenure','MonthlyCharges','TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler() 

data2[scale_col] = scale.fit_transform(data2[scale_col])
```

```python
X = data2.drop('Churn',axis='columns')
Y = data2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
for col in X_train:
    print(f'{col} :{data2[col].unique()}')
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
  tf.keras.layers.Dense(20, input_shape=(26,), activation='sigmoid'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=10)
```

```python
model.evaluate(X_test,Y_test)
```
